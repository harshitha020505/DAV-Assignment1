```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('C:/Users/Harshitha/Downloads/archive (4)/medals.csv')
 # or .json/.xlsx

print(df.head())         # First 5 rows
print(df.tail(3))        # Last 3 rows
print(df.info())         # Info on datatypes and nulls
print(df.isna().sum())   # Missing values
```

```python
df.ffill(inplace=True)  # OR
df['medal_code'] = df['medal_code'].fillna(df['medal_code'].median())
print(df)
```

```python
df['medal_date'] = pd.to_datetime(df['medal_date']) 
print(df)
```

```python
df.drop(columns=['event', 'discipline'], inplace=True)
print(df)
```

```python
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1})
print(df)
```

```python
print("Mean:", df['medal_code'].mean())
print("Median:", df['medal_code'].median())
print("Std Dev:", df['medal_code'].std())
print(df.describe())  # All basic stats
```

```python
filtered_df = df[df['medal_code'] > 2]
sorted_df = df.sort_values(by='medal_code', ascending=False)
print(filtered_df)
print(sorted_df)
```

```python
df['gender1'] = df['gender']
print(df)
```

```python
grouped = df.groupby('medal_code')['code'].count()
print(grouped)
```

```python
pivot1 = pd.pivot_table(df, 
                        index='country', 
                        columns='medal_type', 
                        aggfunc='size', 
                        fill_value=0)
print(pivot1)
```

```python
pivot2 = pd.pivot_table(df, 
                        index='medal_code', 
                        columns='country', 
                        values='medal_type', 
                        aggfunc='count', 
                        fill_value=0)
print(pivot2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Medals by Country
plt.figure(figsize=(12, 6))
df['country'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries by Medal Count')
plt.ylabel('Number of Medals')
plt.xticks(rotation=45)
plt.show()
```

```python
# Medal distribution by type
sns.countplot(data=df, x='medal_type', hue='medal_type', palette='Set2', legend=False)

plt.title('Medal Distribution by Type')
plt.show()
```

```python
df.groupby(['country', 'medal_type']).size().unstack(fill_value=0)
```

```python
df.groupby(['medal_code', 'country'])['medal_type'].count().unstack(fill_value=0)
```

```python
pd.pivot_table(df, index='country', columns='medal_type', aggfunc='size', fill_value=0)
```
